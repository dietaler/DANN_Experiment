```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# 範例資料
categories    = ['feat0', 'feat1', 'feat2']
values_before = [0.223,    0.153,    0.330]
values_after  = [0.35,    0.40,    0.98]

x = np.arange(len(categories))
width = 0.3

plt.figure()
# 先畫 before（設定 alpha=0.5）
plt.bar(x, values_before, width,
        alpha=0.5, label='Before DANN', color='C0', zorder=2)
# 再畫 after（alpha=1.0）
plt.bar(x, values_after,  width,
        alpha=1.0, label='After DANN',  color='C1', zorder=1)

plt.xticks(x, categories)
# plt.xlabel('Feature dimension')
plt.ylabel('accuracy')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
categories    = ['train on 2017\ntest on 2017', 'train on 2017\ntest on 2018', 'train on 2018\ntest on 2018', 'train on 2018\ntest on 2017']
values_beforeDANN = [0.9634, 0.0443, 0.9589, 0.0419]
values_afterDANN  = [0.9493, 0.4005, 0.9131, 0.4525]

x = np.arange(len(categories))
width = 0.3
n = len(categories)

df = pd.DataFrame({
    'category': categories * 2,
    'value':      values_beforeDANN + values_afterDANN,
    'condition': ['without DANN'] * n + ['with DANN'] * n
})

fig, ax = plt.subplots(figsize=(6,4))
sns.barplot(
    x='category', y='value', hue='condition',
    data=df, ci=None, palette=['C0', 'C1'] , alpha=0.8,
    ax=ax
)

for p in ax.patches:
    height = p.get_height()                     # bar 的高度
    ax.annotate(
        f'{height:.4f}',                        # 標籤文字 (4 位小數)
        xy=(p.get_x() + p.get_width()/2, height),  # 文字位置：bar 中心、頂端
        xytext=(0, 3),                            # 向上偏移 3 points
        textcoords='offset points',
        ha='center', va='bottom'
    )
    
# 留空間顯示標籤
ymax = max(values_beforeDANN + values_afterDANN)
ax.set_ylim(0, ymax * 1.10)

# ---- 在這裡重新設定 legend ----
leg = ax.legend(
    loc='upper right',      # 放在右上
    fontsize='small',       # 縮小文字
    title='condition',      # 預設的 title 用同樣大小
    title_fontsize='small',
    markerscale=0.6,        # 縮小色塊
    handletextpad=0.3,      # 色塊與文字間距
    labelspacing=0.2,       # 每行文字間距
    borderpad=0.2,          # legend 外框 padding
    frameon=False           # 去掉外框（選填）
)

plt.ylabel('Accuracy')
plt.xlabel('Tabular Data Type')
plt.tight_layout()
plt.show()
```
